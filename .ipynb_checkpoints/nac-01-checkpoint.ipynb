### **R1**

```python
%matplotlib inline

import cv2
from matplotlib import pyplot as plt
import numpy as np

def circle_transformation(mask_param):
  contorno, _ = cv2.findContours(mask_param, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  mask_param_rgb = cv2.cvtColor(mask_param, cv2.COLOR_GRAY2RGB)

  big = None
  small = None
  small_area = cv2.contourArea(contorno[0])
  big_area = 0

  for item in contorno:
    area = cv2.contourArea(item)
    if area <= small_area:
      small_area = area
      small = item

  for item in contorno:
      area = cv2.contourArea(item)
      if area > big_area:
          big_area = area
          big = item
        
  mask = cv2.fillPoly(mask_param_rgb, [small], 0)
  
  moments = cv2.moments(big)
  if moments["m00"] != 0:
    cx = int(moments["m10"] / moments["m00"])
    cy = int(moments["m01"] / moments["m00"])

  size_cross = 8
  color_cross = (0, 0, 0)
  cv2.line(mask, (cx - size_cross, cy),(cx + size_cross, cy), color_cross, 5)
  cv2.line(mask, (cx, cy - size_cross),(cx, cy + size_cross), color_cross, 5)

  cv2.putText(mask, f'Area is: {str(big_area)}', (cx - 140, cy + 175), cv2.FONT_HERSHEY_SIMPLEX, 1 ,(255, 255, 255),2 , cv2.LINE_AA)

  return mask

img_nac = cv2.imread("imagem-nac.png")
img_hsv = cv2.cvtColor(img_nac, cv2.COLOR_BGR2HSV)

left_circle_lower_hsv = np.array([85, 100, 100])
left_circle_upper_hsv = np.array([100, 255, 255])
right_circle_lower_hsv = np.array([0, 80, 85])
right_circle_upper_hsv = np.array([15, 255, 255])

left_circle_mask = cv2.inRange(img_hsv, left_circle_lower_hsv, left_circle_upper_hsv)
right_circle_mask = cv2.inRange(img_hsv, right_circle_lower_hsv, right_circle_upper_hsv)

left_circle_transformation = circle_transformation(left_circle_mask)
right_circle_transformation = circle_transformation(right_circle_mask)
mask_nac = cv2.bitwise_or(left_circle_transformation, right_circle_transformation)

fig = plt.figure(figsize = (15, 15))
plt.subplot(1, 2, 1)
plt.imshow(mask_nac, cmap = "Greys_r", vmin = 0, vmax = 255)
plt.show()
```

## **R2**

```python
%matplotlib inline

import cv2
from matplotlib import pyplot as plt
import numpy as np
import math

def circle_transformation(mask_param):
  contorno, _ = cv2.findContours(mask_param, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  mask_param_rgb = cv2.cvtColor(mask_param, cv2.COLOR_GRAY2RGB)

  big = None
  small = None
  small_area = cv2.contourArea(contorno[0])
  big_area = 0

  for item in contorno:
    area = cv2.contourArea(item)
    if area <= small_area:
      small_area = area
      small = item

  for item in contorno:
      area = cv2.contourArea(item)
      if area > big_area:
          big_area = area
          big = item
        
  mask = cv2.fillPoly(mask_param_rgb, [small], 0)
  
  moments = cv2.moments(big)
  if moments["m00"] != 0:
    cx = int(moments["m10"] / moments["m00"])
    cy = int(moments["m01"] / moments["m00"])

  return mask, cx, cy

img = cv2.imread("imagem-nac.png")
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

left_circle_lower_hsv = np.array([85, 100, 100])
left_circle_upper_hsv = np.array([100, 255, 255])
right_circle_lower_hsv = np.array([0, 80, 85])
right_circle_upper_hsv = np.array([15, 255, 255])

left_circle_mask = cv2.inRange(img_hsv, left_circle_lower_hsv, left_circle_upper_hsv)
right_circle_mask = cv2.inRange(img_hsv, right_circle_lower_hsv, right_circle_upper_hsv)

mask_left_circle, cx_left_circle, cy_left_circle = circle_transformation(left_circle_mask)
mask_right_circle, cx_right_circle, cy_right_circle = circle_transformation(right_circle_mask)
mask = cv2.bitwise_or(mask_left_circle, mask_right_circle)

base_image = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(base_image, 50, 150)
edges = cv2.cvtColor(canny, cv2.COLOR_GRAY2RGB)
final_img = edges

cv2.line(final_img, (cx_right_circle, cy_right_circle), (cx_left_circle, cy_left_circle), (255, 255, 255), thickness = 1, lineType = 1)   
ang = round(math.degrees(math.atan2(cy_right_circle - cy_left_circle, cx_right_circle - cx_left_circle))) 

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(final_img, f'The angle is: {str(ang)} degrees', (340, 150), font, 1.5, (255, 255, 255), 5, cv2.LINE_AA)

fig = plt.figure(figsize = (15, 15))
plt.subplot(1, 2, 1)
plt.imshow(final_img, cmap = "Greys_r", vmin = 0, vmax = 255)
plt.show()
```
